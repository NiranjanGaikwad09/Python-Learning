# Try - Except - Finally

```python
try:
    print('Try Block')
    try:
        x = {2 , 3 , 4 , 5 ,4 }
        y = [2,3,4,5,6]
        x.update(y)
        print(x)
    except:
        print(traceback.print_exc())
        print("Nested Except Block")
except:
    print('Error Found In - Try Block')
    print('Except Block')
finally:
    print("Finally Block")
```

```python
try:
    print("Try Block")
    print("Hello")
    try:
        x = [2,3,4,4,5]
        y = {2,3,4,5,6}
        x.update(y)    # Error : 'list' object has no attribute 'update'
        print(x)          
        
    except:
        print(traceback.print_exc())
        print("Nested except Block")
        
except:
    print("Error found in Try Block")
    print("Except Block")

finally:
    print("finally Block")
```

## Custom Exceptions in Python (Using raise)

A custom exception is a user-defined error.

You create your own exception class when built-in exceptions are not enough.

Used in:

1 Banking systems

2 Login systems

3 Age validation

4 Business logic rules

# How to Create

1. Create a Class

```python
# It must inherit from Exception.

class MyError(Exception):
    pass
```

 2: Use raise

```python
raise MyError("Something went wrong")
```

### EG - Age Validation

```python
class Age_Error(Exception):
    pass

def Age_Check(age):
    if age < 18:
        raise Age_Error('Cant Enter ! You Are Underage')
    return 'Can Enter'

Age_Check(10)
```

## Using raise with try-except

```python
class Balance_Error(Exception):
    pass

try:
    bal = 1000
    wid = 500
    if wid>bal:
        raise Balance_Error('Insufficient funds')
    print(bal-wid)
    
except Balance_Error as e :
    print("Error : " , e)

```
