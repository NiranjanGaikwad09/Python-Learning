# Loop Control Statements in Python

## 1ï¸âƒ£ break â€” Stop the loop completely

```python
for i in range(1 , 10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(1 , 5):
    if i % 2 == 0:
        break
    print(i)
```

```python
data = [1 , 30 , 67 , -364 , -88]

for i in data :
    if i < 0 :
        print("Invalid data Found")
        break
```

## 2ï¸âƒ£ continue â€” Skip current iteration

```python
for i in range(1 , 10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(1 , 10):
    if i % 2 != 0:
        continue
    print(i)
```

```python
for i in range (1 , 10):
    if i == 3 or i == 6 or i == 9:
        continue
    print(i)
```

```python
data = [10, None, 20, None, 30]

for value in data:
    if value is None:
        continue
    print(value)
```

```python
data = [10 , 20 , True , False , 21]

for i in data :
    if i is True or i is False:
        continue
    print(i)

# >> continue skips the current iteration and moves to the next loop cycle.
```

## 3ï¸âƒ£ pass â€” Do nothing (placeholder)

```python
for i in range(0 , 10):
    if i % 2 == 0 :
        pass
    else:
        print(i)
```
