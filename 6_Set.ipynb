```python
l1 = [1 , 2 , 3 , 4 , 5]
l2 = [1 , 2 , 5 , 6 , 7]

newList = []
# new_list = new list made out of 'l1' items which are in 'l2'


for i in l1:
    if i in l2:
        newList.append(i)

print(newList)
```

```python
l1 = [3,4,5,2,4,7]
l2 = [4,5,6,7,8,9]

new_list = []

for i in l1:
    if i in l2 and i not in new_list:
        new_list.append(i)

print(new_list)
```
