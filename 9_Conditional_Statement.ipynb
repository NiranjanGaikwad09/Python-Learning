## If Else

```python
a = 10 
b = 20 

if a > b :
    print("A is larger")

else: 
    print("B is larger")
```

```python
a = 10 
b = 20 

if a > b :
    print("A is large")

elif a == b :
    print("A and B are equal")

else: 
    print("B is large")
```

```python
a = 20
b = 10

if a > b:  # True
    print("a is greater than b")
    
elif a == b: 
    print("a is equal to b")
    
else:
    print("a is less than b")
```

```python
a , b = 10 , 30

if a == b :
    print("Equal")

if a > b :
    print("A")

if a < b :
    print("B")

if b / a == 3:
    print('a / b = 3')
```

## Examples

### If

```python
age = 21 
if age > 18:
    print("adult")

# Executes only if condition is True
```

### if else

```python
no = 11

if no%2 == 0 :
    print("even")

else:
    print("odd")
```

```python
marks = 80

if marks >= 90 :
    print("A")
elif marks >= 75:
    print("B")

elif marks >=40:
    print("pass")

else:
    print("FAil")
```

### Nested if

```python
age = 21 
id = False

if age > 18:
    if id == True:
        print("Can Vote")
    else:
        print("Cannot")
```

### Ternary (One-line if) ------ Short form of if–else

```python
a = 10 
b = 23

print("A is greater than b ") if a > b else print("B is greater")
```

### Logical operators with condition`

```python
age = 21 
citizen = True

if age > 18 and citizen == True:
    print("Can Vote")
```
