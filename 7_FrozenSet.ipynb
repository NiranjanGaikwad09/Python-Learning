```python
fs3 = fs1.union(fs2)
```

## intersection()       - Returns Common elements


```python
fs1 = {1 , 3 , 5, 7 , 9}
fs2 = {'idli' , 3.13 , 3+6j , 9 , 3}

fs1.intersection(fs2)    
```

```python
for i in fs1.intersection(fs2):
    if i % 2 == 0:
        print(i)
    else:
        print(f"{i} - is Not div by 2")
```

## difference()   - Unique elements from both set

```python
fs1 = {1 ,  22 , 3 , 5, 7 , 9}
fs2 = {'idli' , 3.13 , 3+6j , 9 , 3}
```

```python
fs1
```

```python
fs2
```

```python
fs1.difference(fs2)
```

```python
cnt = 0
for i in fs1.difference(fs2):
    cnt += 1
    print(f"{cnt} : {i}")
    
```

```python
for i in fs1.difference(fs2):
    if i % 2 == 0 :
        print(f"{i} is : even")
    elif i % 2 != 0 :
         print(f"{i} is : odd")
    else :
        print("NA")
```

## symmetric_difference()


```python
f1 = {1 , 40 , 88 , 11 , 10}
f2 = {7 , 45 , 87 , 34 , 4}

a = f1.symmetric_difference(f2)
print(f1)
print(f2)
print(a)
```

```python
for i in a :
    if i % 2 == 0:
        print(f"{i} : Even")
    elif i % 2 != 0:
        print(f"{i} : Odd")

    else:
        ("NA")
    
```
