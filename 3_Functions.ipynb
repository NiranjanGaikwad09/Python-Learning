## Global And Local Variable 

```python
1. Gloabl Variable 
2. Local Variable 
```

```python
x = 10
y = 20

# The are declared outside the Function - so it global 
```

```python
def get_addition():
    print(x)
    print(y)
    add = x + y 
    print(add)
    
get_addition()
print(x)
```

```python
def get_addition():
    x1 = 12         # Local 
    y2 = 22         # Local

    add1 = x1 + y2
    print(add1)

get_addition()
#print(x1)    
```

```python
def get_addition():
    x1 = 12         # Local 
    y2 = 22         # Local

    add1 = x1 + y2
    print(add1)

get_addition()        # Function ends here 
print(x1)             # Cant access a Local variable outside a function  >>>> name 'x1' is not defined
```

```python
def add_func():
    x1 = 10
    x2 = 12
    
    global add
    add = x1 + x2 
    print(add)

add_func()
print(add)
```

```python
def add_func():
    x1 = 10
    x2 = 12
    global add
    
    add = x1 + x2
    print(add)

add_func()
```
