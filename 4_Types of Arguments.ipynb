```python
def my_func(**kwargs):
    print(f"Kwargs : {kwargs}")
    print("Values : " , kwargs.values())
    res = 100
    return res

my_func(a = 10 , b = 12 , c = 13 , d = 14)
```

```python
def my_func(**kwargs):
    print(f"kwargs >> {kwargs}")
    print(f"sum of values >>> {sum(kwargs.values())}")

    res = 100
    return res

my_func(a = 10 , b = 12 , c = 14 , d = 15)
```

```python
def my_func(**kwargs):
    print("Kwargs >> ", kwargs)
    print("Sub Name : ", list(kwargs.keys()))
    
    print("Marks Obtained : ", list(kwargs.values()))
    percent = sum(kwargs.values())/ len(kwargs)
    
    f_res = f"Percent : {percent}"
    return f_res

my_func(eng = 100 , mar = 90 , math = 95 , sci = 99)
```

```python
def my_func(**kwargs):
    print("Kwargs : ", kwargs)
    print("Sum : ", sum(kwargs.values()))
    percent = sum(kwargs.values()) / len(kwargs.keys())
    return percent

my_func(math = 100 , eng = 89 , mar = 90 , hin = 88)
```
