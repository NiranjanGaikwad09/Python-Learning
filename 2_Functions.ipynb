```python
def Calc(iterable):
    print("Iterable : " , iterable)
    sum1 = 0
    mul1 = 1

    for i in iterable:
        if type(i) == str:
            i = eval(i)

            sum1 += i
            mul1 *= i

    print("Addition is : " , sum1)
    print("Multi is : " , mul1)

Calc([52 , 3 , 1])
```
