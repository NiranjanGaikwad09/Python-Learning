## 5️⃣ *args (Variable Positional Arguments)
Used when you don’t know how many arguments will be passed.

```python
def add(*args):
    tot = 0
    for i in args:
        tot += i
    return tot
    
add(1 , 2 , 4 , 5)

# >> args is a tuple
```

## 6️⃣ **kwargs (Variable Keyword Arguments)
Used for key-value arguments


```python
def std(**kwargs):
    for key, value in kwargs.items():
        print(key ," : " , value )

std(name = 'Nira' , std = 10 , city ="Pune")

# >> kwargs is a dictionary
```

```python
def emp (**kwargs):
    for k , v in kwargs.items():
        print(k , "::" , v)

emp(Name = "Didi" , Company = "Wipro" , Doj = "Oct 2025")
```

## 7️⃣ Using *args and **kwargs together

```python
def demo(*args , **kwargs):
    print(args)
    print(kwargs)

demo(10 , 20 , 22 , a = 100 , b = 120 , c = 21)

# >> parameters → *args → **kwargs
```

```python
def demo1(**kwargs , *args):
    print(args)
    print(kwargs)

demo1(a = 10 , b = 21 , c = 22 , 10 , 21 , 122)

# Error : >>  arguments cannot follow var-keyword argument
```

## 8️⃣ Lambda Functions (Anonymous Functions)
A lambda function is a one-line function without a name

```python
lambda arguments : expression
```

```python
add = lambda a , b : a+b
print(add(10 , 12))
```

```python
num = [1 , 2 , 3 , 4]
sqr =list(map(lambda x : x*x , num))
print(sqr)
```

```python
l1 = range(1 , 10+1)

sqr = lambda l1 : [i**2 for i in l1]
sqr(l1)
```
