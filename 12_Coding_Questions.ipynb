```python
std_name = {'Nira' , 'didi' , 'duddi' , 'onil' , 'nisa'}
pass_std = {'Nira' , 'didi' , 'nini' , 'guddi' }

u_set = std_name.union(pass_std)
print(u_set)
# Union - gives unique items from both set
```

```python
std_name = {'Nira' , 'didi' , 'duddi' , 'onil' , 'nisa'}
pass_std = {'Nira' , 'didi' , 'nini' , 'guddi' , 'onil' }

intersection_set = std_name.intersection(pass_std)

print("intersection : " , intersection_set)
# Intersection >> items present in both set
```

```python
std_name = {'Nira' , 'didi' , 'duddi' , 'onil' , 'nisa'}
pass_std = {'Nira' , 'didi' , 'nini' , 'guddi' , 'onil' }

std_dif = std_name.difference(pass_std)

print("Diff : ", std_dif)

# .difference >> difference() returns elements present in the first set but not in the second
```
