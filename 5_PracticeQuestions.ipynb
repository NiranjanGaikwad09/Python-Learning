```python
a = 'NiraNjan'

upp_a = ''

for ch in a:
    if 'a' <= ch <= 'z':
        upp_a += chr(ord(ch)-32)

    else:
        upp_a += ch

print(upp_a)
```

```python
s = "hello"
mapping = {chr(i): chr(i - 32) for i in range(ord('a'), ord('z') + 1)}

result = ""
for ch in s:
    if ch in mapping:
        result += mapping[ch]
    else:
        result += ch

print(result)   # HELLO
```

```python
a = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
cnt = {}

for ch in a.items() :
   if ch in cnt:
       cnt[ch] += 1

   else:
       cnt[ch] = 1

print(cnt)
    
```

```python
a = {'a': 10, 'b': 20, 'c': 30, 'd': 40}

cnt = 0

for k  , v  in a.items():
    if isinstance(k , str):
        cnt += 1

print(cnt)
```
