```python
a = 100
b = 'niranjan'

# This Global Variable - Because its not declared inside any function
```

```python
def test_function():
    y = 100
    # Now y is Locat variable - its declared within a function

    sq_num = y ** 2
    return sq_num
```

```python

test_function()
```

```python
print(y)
#While printing y its giving error - cant access local variable
# -- NameError: name 'y' is not defined
```

```python
def test_function():
    global y 
    y = 100

    sq_num = y ** 2
    return sq_num
```

```python
print(y)
# Now we can access y -Because its set to gloabl 

test_function()
```
