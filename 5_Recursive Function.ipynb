## Factorial 

```python
def Fact(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
        
    print(f"Factorial of {n} is : {fact}")
    return fact

Fact(5)
```

```python
def fact(n):
    if n == 1:
        return 1 
    else :
        fact = 1
        for i in range (1 , n+1):
            fact *= i
        return fact

fact(5)
```

```python
def fact(n):
    if n == 1:
        return 1
    else:
        return n * fact(n - 1)     # 5 * fact(5-1)
                                   # 5 * 4 * fact(4-1)
                                   # 5 * 4 * 3 * fact(3-1)
                                   # 5 * 4 * 3 * 2 * fact(2-1)
n = 5
fact(n)
```

```python
def fact(n):
    if n ==1 :
        return 1 
    else :
        return n * fact(n-1)

fact(6)
```

# 5ï¸âƒ£ Classic Example: Factorial

```python
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

fact(5)

# Flow

# 5*           fact(4)
# 5*4*         fact(3)
# 5*4*3*       fact(2)
# 5*4*3*2*     fact(1)
# 5*4*3*2*1
```

# 7ï¸âƒ£ Another Important Example: Fibonacci

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)

print(fib(90))
```
