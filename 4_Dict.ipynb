```python
EMP_DETAILS = {'Name' : 'Shraddha' , 'Age' : 25 , 'Job Role' : 'Ai Eng' , 'Salary' : 20 }

NEW_entry = {'Location' : 'Pune' , 'Weight' : 60}

EMP_DETAILS.update(NEW_entry)
EMP_DETAILS
```

```python
LIST = [1 ,2 ,3 ,4 ,5 ,6]

SQR = {}

for i in LIST :
    SQR.update({i : i**2})
    
print(SQR)
```

```python
l = [5 , 10 , 15 , 20]
sqr = {}

for i in l :
    sqr.update({i : i + 9})

print(sqr)
```
